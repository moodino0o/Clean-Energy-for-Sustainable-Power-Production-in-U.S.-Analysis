```python
# replacing the comma with empty string so it can be converted from object to int
def comma_replace (table):
        return str(table).replace(',','')
def dot_replace (table):
        return str(table).split('.')[0]
```

### Melting

```python
netgen = netgen_df.melt(
    id_vars=["index",'YEAR'],
    value_vars=Netgen_col,
    var_name="Month",
    value_name="NetGen"
)
netgen["Month"] = netgen["Month"].str.replace("Netgen ", "", regex=False)
netgen["Date"] = netgen["YEAR"].astype(str) + " " + netgen["Month"]
```

```python
netgen
```

```python
Quantity = Quantity_df.melt(
    id_vars=["index",'YEAR'],
    value_vars=Quantity_col,
    var_name="Month",
    value_name="Quantity"
)

Quantity["Month"] = Quantity["Month"].str.replace("Quantity", "", regex=False)
Quantity["Date"] = Quantity["YEAR"].astype(str) + " " + Quantity["Month"]
```

```python
Elec_Quantity = Elec_Quantity_df.melt(
    id_vars=["index",'YEAR'],
    value_vars=Elec_Quantity_col,
    var_name="Month",
    value_name="Elec_Quantity"
)

Elec_Quantity["Month"] = Elec_Quantity["Month"].str.replace("Elec_Quantity", "", regex=False)
Elec_Quantity["Date"] = Elec_Quantity["YEAR"].astype(str) + " " + Elec_Quantity["Month"]
```

```python
MMBtuPer = MMBtuPer_df.melt(
    id_vars=["index",'YEAR'],
    value_vars=MMBtuPer_col,
    var_name="Month",
    value_name="MMBtuPer"
)

MMBtuPer["Month"] = MMBtuPer["Month"].str.replace("MMBtuPer_Unit", "", regex=False)
MMBtuPer["Date"] = MMBtuPer["YEAR"].astype(str) + " " + MMBtuPer["Month"]
```

```python
Tot_MMBtu = Tot_MMBtu_df.melt(
    id_vars=["index",'YEAR'],
    value_vars=Tot_MMBtu_col,
    var_name="Month",
    value_name="Tot_MMBtu"
)

Tot_MMBtu["Month"] = Tot_MMBtu["Month"].str.replace("Tot_MMBtu", "", regex=False)
Tot_MMBtu["Date"] = Tot_MMBtu["YEAR"].astype(str) + " " + Tot_MMBtu["Month"]
```

```python
Elec_MMBtu = Elec_MMBtu_df.melt(
    id_vars=["index",'YEAR'],
    value_vars=Elec_MMBtu_col,
    var_name="Month",
    value_name="Elec_MMBtu"
)

Elec_MMBtu["Month"] = Elec_MMBtu["Month"].str.replace("Elec_MMBtu", "", regex=False)
Elec_MMBtu["Date"] = Elec_MMBtu["YEAR"].astype(str) + " " + Elec_MMBtu["Month"]
```
